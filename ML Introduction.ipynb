

## üîπ Artificial Intelligence (AI)

### **Definition**

* AI is the **broad field** of computer science that focuses on creating machines that can **mimic human intelligence**.
* The ultimate goal: make machines capable of **reasoning, decision-making, problem-solving, perception, and learning**.

### **Key Idea**

* Instead of programming every step ‚Üí AI systems can adapt and decide what to do.

### **Examples**

* **Self-driving cars**: Recognize pedestrians, traffic signs, make driving decisions.
* **Virtual assistants**: Siri, Alexa, Google Assistant ‚Üí understand voice, respond.
* **Robotics**: Industrial robots optimizing workflows.

üëâ Think of **AI as the parent field** that includes ML, DL, and other methods like expert systems, rule-based systems, search algorithms.

## üîπ Machine Learning (ML)

### **Definition**

* ML is a **subset of AI**.
* Machines **learn from data** by identifying patterns and improving with experience ‚Äî without needing explicit programming for every rule.

### **How It Works**

* Input: Historical data.
* Process: Algorithms detect relationships/patterns.
* Output: Predictions or classifications.

### **Examples**

* **Spam detection** in Gmail ‚Üí learns from labeled examples (spam vs. not spam).
* **House price prediction** ‚Üí learns relationship between size, location, price.
* **Fraud detection** ‚Üí identifies unusual transaction patterns.

### **Key Difference vs AI**

* AI = intelligence broadly.
* ML = the way machines **achieve intelligence** by learning from data.

---

## üîπ Deep Learning (DL)

### **Definition**

* DL is a **subset of ML**.
* Uses **Artificial Neural Networks (ANNs)** with many hidden layers (‚Äúdeep‚Äù layers).
* Inspired by how the human brain processes information through neurons.

### **Why Special?**

* Unlike traditional ML, DL can **automatically extract features** from raw data.
* Handles **complex, unstructured data** like images, video, audio, and natural language.

### **Examples**

* **Face recognition** on Facebook/phones.
* **Voice-to-text** in Alexa, Google Voice.
* **Autonomous driving**: Detecting objects, lanes, pedestrians.

üëâ DL is what made breakthroughs like **computer vision and speech recognition** possible.

## üîπ Data Science

### **Definition**

* Data Science is an **interdisciplinary field** that combines:

  1. **Mathematics & Statistics** ‚Üí understanding data.
  2. **Programming** ‚Üí handling & analyzing data.
  3. **ML/DL** ‚Üí building predictive models.
  4. **Domain knowledge** ‚Üí applying results to real-world problems.

### **What Data Scientists Do**

* Collect and clean data.
* Explore data with statistics & visualization.
* Apply ML/DL algorithms.
* Interpret results and **communicate insights** to stakeholders.

### **Examples**

* **Netflix recommendations**: Suggest shows based on viewing history.
* **Amazon product suggestions**: Predict what users might buy.
* **Business forecasting**: Predicting sales, demand, customer churn.

* **DL** = A special type of ML (neural networks with depth).
* **Data Science** = Applies ML/DL + stats + business knowledge to solve real-world problems.
