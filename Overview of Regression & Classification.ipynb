

## 🔹 1. What are ML Tasks?

* In **Supervised Learning**, we give the algorithm both **input (X)** and **output (y)**.
* The model learns the mapping **X → y**.
* Two main supervised tasks are: **Regression** and **Classification**.

---

## 🔹 2. Regression

### **Definition**

* Regression is used when the **output (y) is continuous**.
* Goal: Predict a **numerical value**.

### **Examples**

* Predicting **house price** from size, location, rooms.
* Predicting **salary** based on years of experience.
* Predicting **temperature** for tomorrow.

### **Key Idea**

* The model fits a curve/line that minimizes prediction error.
* Formula for Linear Regression:

  $$
  y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
  $$

👉 Regression = *“How much?”*

---

## 🔹 3. Classification

### **Definition**

* Classification is used when the **output (y) is categorical/discrete**.
* Goal: Predict a **class label**.

### **Examples**

* Spam (1) vs. Not Spam (0) emails.
* Disease (Yes/No) prediction.
* Handwritten digit recognition (0–9).

### **Key Idea**

* The model learns decision boundaries between classes.
* Logistic Regression uses the **Sigmoid function** to map outputs to probability:

  $$
  P(y=1|x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \dots)}}
  $$

👉 Classification = *“Which category?”*

---

## 🔹 4. Difference Between Regression and Classification

| Aspect            | Regression                       | Classification                  |
| ----------------- | -------------------------------- | ------------------------------- |
| **Output**        | Continuous (numbers)             | Discrete (categories/labels)    |
| **Examples**      | House price, salary, temperature | Spam detection, disease, digits |
| **Question Type** | “How much?”                      | “Which class?”                  |

---

## 🔹 5. summarize

* In short :

  * “If your output is a continuous variable, it’s a regression problem.”
  * “If your output is a category, it’s a classification problem.”

---

✅ clearly understand:

* ML tasks are divided mainly into **Regression vs Classification**.
* Regression = predicting numbers.
* Classification = predicting categories.
