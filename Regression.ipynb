
## üîπ 1. What is Linear Regression?

* **Definition**: Linear Regression is a **supervised learning algorithm** used for **regression problems**.
* It models the relationship between **independent variables (X)** and a **dependent variable (y)** using a straight line.
* Equation of a simple linear regression:

$$
y = \beta_0 + \beta_1 x + \epsilon
$$

where:

* $y$ = dependent variable (output)
* $x$ = independent variable (input)
* $\beta_0$ = intercept (bias)
* $\beta_1$ = slope (coefficient/weight)
* $\epsilon$ = error term

üëâ Idea: Find the **best fit line** that minimizes prediction errors.

---

## üîπ 2. Cost Function (Error Measurement)

* To evaluate how well the line fits, we use **Mean Squared Error (MSE):**

$$
J(\beta_0, \beta_1) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

where $\hat{y}_i$ = predicted value.
üëâ Goal: Minimize the cost function.

---

## üîπ 3. Optimization (How We Learn Coefficients)

* The model finds coefficients ($\beta$) using methods like:

  * **Ordinary Least Squares (OLS)** (exact solution using linear algebra).
  * **Gradient Descent** (iterative optimization for large datasets).

---

## üîπ 4. Assumptions of Linear Regression

1. **Linearity**: Relationship between X and y is linear.
2. **Independence**: Observations are independent.
3. **Homoscedasticity**: Equal variance of errors.
4. **Normality**: Errors should follow normal distribution.
5. **No Multicollinearity**: Independent variables should not be highly correlated.

---

## üîπ 5. Pros & Cons

### ‚úÖ Pros

* Simple and easy to understand.
* Fast to train, interpretable (coefficients show importance).
* Good for small/medium datasets.

### ‚ùå Cons

* Works only for linear relationships.
* Sensitive to outliers.
* Assumes independence and normality of errors.
* Limited in capturing complex patterns.

---

## üîπ 6. When to Use

* When you want **interpretability**.
* When data shows **linear trend**.
* When prediction accuracy is less critical than **understanding relationships**.

---

## üîπ 7. Example 

### Example Problem: Predict House Price

* Input (X): Size of house (sq. ft.)
* Output (y): House Price

Steps:

1. Collect data: (1000 sq ft ‚Üí 50 lakhs, 1500 sq ft ‚Üí 70 lakhs, etc.).
2. Fit a straight line using OLS or Gradient Descent.
3. Predict for new input (e.g., 1200 sq ft).

---

## üîπ 8. Hands-On Code (Simple Linear Regression)




```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data (House Size vs Price)
X = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  # feature
y = np.array([50, 70, 90, 110, 130])  # target (in lakhs)
```

```python
# Train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predictions
y_pred = model.predict(X)
```

```python
# Print coefficients
print("Intercept (Œ≤0):", model.intercept_)
print("Slope (Œ≤1):", model.coef_)
```

```python
# Plot
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(X, y_pred, color="red", label="Best Fit Line")
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price (lakhs)")
plt.legend()
plt.show()
```


---

### Output (Example):

* Intercept: 10
* Slope: 0.04
* Equation: $y = 10 + 0.04x$
  
  üëâ Means: For every +100 sq ft, price increases by 4 lakhs.

# üìå Multiple Linear Regression (MLR)

---

## üîπ 1. Definition

* **Multiple Linear Regression** is an extension of **Simple Linear Regression** where we use **two or more independent variables** to predict a dependent variable.
* Equation:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon
$$

where:

* $y$ = dependent variable
* $x_1, x_2, ..., x_n$ = independent variables
* $\beta_0$ = intercept
* $\beta_1, \beta_2, ..., \beta_n$ = coefficients (slopes)
* $\epsilon$ = error

üëâ Instead of fitting a line in 2D, we fit a **hyperplane** in n-dimensional space.

---

## üîπ 2. Example Problem

**House Price Prediction**

* Dependent Variable (y): Price of house.
* Independent Variables (X):

  * $x_1$ = size (sq. ft.)
  * $x_2$ = number of bedrooms
  * $x_3$ = distance from city center
  * $x_4$ = age of house

Equation might look like:

$$
\text{Price} = 5 + 0.04(\text{Size}) + 10(\text{Bedrooms}) - 3(\text{Age}) - 2(\text{Distance})
$$

---

## üîπ 3. How It Works

1. Collect training data with multiple predictors.
2. Fit a hyperplane that minimizes **Mean Squared Error (MSE)**.
3. Use coefficients ($\beta$) to interpret effect of each predictor.

---

## üîπ 4. Assumptions of MLR

* Same as Linear Regression **+ Multicollinearity check**.

1. **Linearity**: Relationship between predictors and target is linear.
2. **Independence** of errors.
3. **Homoscedasticity** (constant variance).
4. **Normality** of errors.
5. **No Multicollinearity**: Independent variables should not be highly correlated.

---

## üîπ 5. Pros & Cons

### ‚úÖ Pros

* Can handle multiple factors simultaneously.
* Provides insights into variable importance.
* Easy to implement and interpret.

### ‚ùå Cons

* Assumes linear relationship.
* Multicollinearity can mislead results.
* Sensitive to outliers.
* Cannot model complex non-linear patterns.

---

## üîπ 6. When to Use

* When you want to analyze the **impact of multiple variables** on an outcome.
* When relationships are approximately linear.
* For predictive tasks with tabular data.

---

## üîπ 7. Hands-On Example (Sklearn)



```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Sample dataset
data = {
    "Size_sqft": [1000, 1500, 2000, 2500, 3000],
    "Bedrooms": [2, 3, 3, 4, 4],
    "Age": [10, 5, 8, 2, 1],
    "Price_lakhs": [50, 70, 90, 110, 130]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Features (X) and Target (y)
X = df[["Size_sqft", "Bedrooms", "Age"]]
y = df["Price_lakhs"]
```

```python
# Train Model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predictions
pred = model.predict(X)
```

```python
# Coefficients
print("Intercept (Œ≤0):", model.intercept_)
print("Coefficients (Œ≤1, Œ≤2, Œ≤3):", model.coef_)
```

```python
import matplotlib.pyplot as plt

# Actual vs Predicted plot
plt.scatter(y, pred, color="blue")
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)  # reference line
plt.xlabel("Actual Price (Lakhs)")
plt.ylabel("Predicted Price (Lakhs)")
plt.title("Actual vs Predicted Prices - Multiple Linear Regression")
plt.show()
```

### Possible Output

* Intercept: 5.2
* Coefficients: \[0.04, 8.5, -1.2]

Equation becomes:

$$
\text{Price} = 5.2 + 0.04(\text{Size}) + 8.5(\text{Bedrooms}) - 1.2(\text{Age})
$$

üëâ Interpretation:

* Every **extra sq ft** increases price by 0.04 lakhs.
* Each **additional bedroom** increases price by 8.5 lakhs.
* Every **extra year of age** reduces price by 1.2 lakhs.
